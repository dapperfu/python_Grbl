## Jupyter Notebooks Integration

Type out g-code / grbl commands directly into a notebook

```python
from IPython.core.magic import register_cell_magic

@register_cell_magic
def grbl_cell_magic(cfg, cell):
    "# Sending G-Code"
    grbl.run(cell)
```
